# Explanation of Affine Transformation Example:
imread  access to the image,0 change color to grey.

img.shape==> use this to accessing Image Properties that return truple of numbers of rows, columns and channels (if image is color) But if image is grayscale tuple returned contains only number of rows and columns.in this example the image is gray so returm 2 attributs .

ask about float32 function.

getAffineTransform()===> this function use to Get the Affine Transform.

warpAffine(orginal image,new image ,(num of cols in the orginal image,num of  rows in tne orginal image))==> apply the Affine Transform just found to the src image. ??? 

subplot(121 OR 122)===> Specifies the location of the output image.

title('title')===> print title above the image .

show()==>to View the output on the plt order.





```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread('images/drawing.png')
rows,cols,ch = img.shape
pts1 = np.float32([[50,50],[200,50],[50,200]])
pts2 = np.float32([[10,100],[200,50],[100,250]])
M = cv2.getAffineTransform(pts1,pts2)
dst = cv2.warpAffine(img,M,(cols,rows))
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```
