# Average Filtering :
what is the Average Filter?
Average filtering (linear )is a method of ‘smoothing’ images by
reducing the amount of intensity variation between neighbouring
pixels.

how the Average Filter work?
The average filter works by moving through the image pixel by pixel,
replacing each value with the average value of neighbouring pixels,
including itself.

# Explanation of  Average Filtering Example:\
ones(shape,datatype)===>
*shape:int or sequence of ints Shape of the new array
*datatype : data-type, optional
this function use to return a new array of given shape and type, filled with ones.


filter2D(Orginal image,ddepth,kernal)===>Convolves an image with the kernel.
*ddepth: desired depth of the destination image; if it is negative, it will be the same as src.depth().
*Kernal: Represent the mask.

subplot(121 OR 122)===> Specifies the location of the output image.

title('title')===> print title above the image .

xticks([])===> To remove from the division in the axis of the X.

yticks([])===> To remove from the splitter in the axis of the Y.

show()==> To View the output on the plt order.




```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('images/nn.jpg')

kernel = np.ones((9,9),np.float32)/81
dst = cv2.filter2D(img,-1,kernel)

plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('Averaging')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('images/nn.jpg')

blur = cv2.blur(img,(5,5))

plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```
