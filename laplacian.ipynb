# laplacian

laplacian() function it defined the frame. of object in the picture and it know the object by 2 way the first is by color second way by continuous edges.




```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('image/home.jpg',0)

laplacian = cv2.Laplacian(img,cv2.CV_64F)
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=21)
sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=21)

plt.subplot(2,2,1),plt.imshow(img,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,2),plt.imshow(laplacian,cmap = 'gray')
plt.title('Laplacian'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,3),plt.imshow(sobelx,cmap = 'gray')
plt.title('Sobel X'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,4),plt.imshow(sobely,cmap = 'gray')
plt.title('Sobel Y'), plt.xticks([]), plt.yticks([])

plt.show()
```

# Add Orginal img to Laplacian img 
when i add the orginal img to laplacian the result image display the edges more than orginal image ,view the object frame mare clearly than other images .

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('image/home.jpg',0)

laplacian = cv2.Laplacian(img,cv2.CV_64F)
addimg= np.add(img,laplacian)

plt.subplot(2,2,1),plt.imshow(img,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,2),plt.imshow(laplacian,cmap = 'gray')
plt.title('Laplacian'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,3),plt.imshow(addimg,cmap = 'gray')
plt.title('Add'), plt.xticks([]), plt.yticks([])


plt.show()
```

when i use subtract function i see the different result Compare with the result of the add image i think the different is in sub image the object frame is more and more clearly than add image.

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('image/home.jpg',0)

laplacian = cv2.Laplacian(img,cv2.CV_64F)
subimg= np.subtract(img,laplacian)
addimg= np.add(img,laplacian)
plt.subplot(2,2,1),plt.imshow(img,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,2),plt.imshow(laplacian,cmap = 'gray')
plt.title('Laplacian'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,3),plt.imshow(subimg,cmap = 'gray')
plt.title('Sub'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,4),plt.imshow(addimg,cmap = 'gray')
plt.title('Add'), plt.xticks([]), plt.yticks([])


plt.show()
```
