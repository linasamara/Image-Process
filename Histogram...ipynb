# after the histogram.
i want do threshold 
threshold(gray image,165,cv2.THRESH_BINARY)===>do threshold to do image black and white color ret must put that before the theshold attributes. the object white and the background black.


bitwise_and(gray image, The result of the thresh) ====> to multiply gray image and The result of the thresh to show the object gray and the background black.

after that i do threshold on gray image but i use THRESH_BINARY_INV this to show the background white and the object black. here Reflective result first theshold.\

do bitwise_and on second threshold  to show the background gray.after that i do Gaussian Blur to blur background .

add(The image on which the object appeared is gray  , result from blur)===>Merge the two images


```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
gimg = cv2.imread('images/f.jpg',0)
oimg = cv2.imread('images/f.jpg')
hist=plt.hist(gimg.ravel(),256,[0,256])#here by the matplotlib to do histogram
plt.title('Histogram')
plt.show()
cv2.destroyAllWindows()
ret,thresh1 = cv2.threshold(gimg,165,cv2.THRESH_BINARY) #do threshold to do image black and white color ret must put that before the theshold att

 
#bimg = cv2.multiply(oimg, (500,750,1))
objectimg = cv2.bitwise_and(gimg, thresh1)#هون بطلع القمر جراي والخلفية اسود
#bimg = oimg * thresh1
ret,BGwhiteimg = cv2.threshold(gimg,127,255,cv2.THRESH_BINARY_INV)
GrayBG = cv2.bitwise_and(gimg,BGwhiteimg)#بطلع الخلفية جراي 
blur = cv2.GaussianBlur(GrayBG,(91,91),0)#بعمل smoth للخلفية 


addimg = cv2.add(objectimg, blur)#دمج للصورتين 
#Colorimg = cv2.cvtColor(addimg,cv2.COLOR_GRAY2RGB)



#cv2.imshow('resize 1', addimg) 
#cv2.waitKey(0)
#cv2.destroyAllWindows()
titleArr = ['Orginal Images','Grey Images','Image threshold','new mimge','White background image','gray background','Blur image','add images']

imagesArr= [oimg,gimg,thresh1,objectimg,BGwhiteimg,GrayBG,blur,addimg]#must do array to print images.

for i in range(8):
   # plt.subplot(2,3,i+1),plt.imshow(imagesArr[i],'grey')
    plt.subplot(3,3,i+1),plt.imshow(imagesArr[i],'gray')
    plt.title(titleArr[i])
    plt.xticks([]), plt.yticks([])
    plt.show()
    
    
    
```
